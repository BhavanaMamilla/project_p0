```python
import mysql.connector
from mysql.connector import Error
try:
    # Establish a connection to the MySQL database
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="Bhavana@123",
        database="revstore"
    )
    if conn.is_connected():
        print("Successfully connected to the database")
    cursor = conn.cursor()
    # Create Customers Table
    try:
        cursor.execute("""
        CREATE TABLE IF NOT EXISTS Customers (
            customer_id INT PRIMARY KEY,
            customer_name VARCHAR(255),
            country VARCHAR(255),
            city VARCHAR(255)
        )
        """)
        print("Customers table created successfully")
    except Error as e:
        print(f"Error creating Customers table: {e}")
    # Create Products Table
    try:
        cursor.execute("""
        CREATE TABLE IF NOT EXISTS Products (
            product_id INT PRIMARY KEY,
            product_name VARCHAR(255),
            product_category VARCHAR(255)
        )
        """)
        print("Products table created successfully")
    except Error as e:
        print(f"Error creating Products table: {e}")
    # Create Payments Table
    try:
        cursor.execute("""
        CREATE TABLE IF NOT EXISTS Payments (
            payment_txn_id VARCHAR(255) PRIMARY KEY,
            payment_type VARCHAR(255),
            payment_txn_success VARCHAR(255),
            failure_reason VARCHAR(255)
        )
        """)
        print("Payments table created successfully")
    except Error as e:
        print(f"Error creating Payments table: {e}")
    # Create Orders Table
    try:
        cursor.execute("""
        CREATE TABLE IF NOT EXISTS Orders (
            order_id INT PRIMARY KEY,
            customer_id INT,
            product_id INT,
            payment_txn_id VARCHAR(255),
            qty INT,
            price DECIMAL(10, 2),
            datetime DATETIME,
            ecommerce_website_name VARCHAR(255),
            FOREIGN KEY (payment_txn_id) REFERENCES Payments(payment_txn_id),
            FOREIGN KEY (product_id) REFERENCES Products(product_id),
            FOREIGN KEY (customer_id) REFERENCES Customers(customer_id)
        )
        """)
        print("Orders table created successfully")
    except Error as e:
        print(f"Error creating Orders table: {e}")
    # Commit the changes
    conn.commit()
except Error as e:
    print(f"Database connection error: {e}")
finally:
    # Ensure the connection is closed even if an error occurs
    if conn.is_connected():
        cursor.close()
        conn.close()
        print("Database connection closed")
```

```python
import mysql.connector
from mysql.connector import Error
try:
    # Establish a connection to the MySQL database
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="Bhavana@123",
        database="revstore"
    )
    if conn.is_connected():
        print("Successfully connected to the database")
    cursor = conn.cursor()
    # Insert into Orders Table
    try:
        cursor.executemany("""
        INSERT INTO Orders (order_id, customer_id, product_id, payment_txn_id, qty, price, datetime, ecommerce_website_name)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        ON DUPLICATE KEY UPDATE customer_id = VALUES(customer_id), product_id = VALUES(product_id), payment_txn_id = VALUES(payment_txn_id), qty = VALUES(qty), price = VALUES(price), datetime = VALUES(datetime), ecommerce_website_name = VALUES(ecommerce_website_name)
        """, df1.values.tolist())
        print("Data inserted into Orders table successfully")
    except Error as e:
        print(f"Error inserting data into Orders table: {e}")
    # Insert into Customers Table
    try:
        cursor.executemany("""
        INSERT INTO Customers (customer_id, customer_name, country, city)
        VALUES (%s, %s, %s, %s)
        ON DUPLICATE KEY UPDATE customer_name = VALUES(customer_name), country = VALUES(country), city = VALUES(city)
        """, dfc.values.tolist())
        print("Data inserted into Customers table successfully")
    except Error as e:
        print(f"Error inserting data into Customers table: {e}")
    # Insert into Products Table
    try:
        cursor.executemany("""
        INSERT INTO Products (product_id, product_name, product_category)
        VALUES (%s, %s, %s)
        ON DUPLICATE KEY UPDATE product_name = VALUES(product_name), product_category = VALUES(product_category)
        """, df2.values.tolist())
        print("Data inserted into Products table successfully")
    except Error as e:
        print(f"Error inserting data into Products table: {e}")
    # Insert into Payments Table
    try:
        cursor.executemany("""
        INSERT INTO Payments (payment_txn_id, payment_type, payment_txn_success, failure_reason)
        VALUES (%s, %s, %s, %s)
        ON DUPLICATE KEY UPDATE payment_type = VALUES(payment_type), payment_txn_success = VALUES(payment_txn_success), failure_reason = VALUES(failure_reason)
        """, df3.values.tolist())
        print("Data inserted into Payments table successfully")
    except Error as e:
        print(f"Error inserting data into Payments table: {e}")
    # Commit the changes
    conn.commit()
except Error as e:
    print(f"Database connection error: {e}")
finally:
    # Ensure the connection is closed even if an error occurs
    if conn.is_connected():
        cursor.close()
        conn.close()
        print("Database connection closed")
```
